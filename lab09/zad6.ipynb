```python
import numpy as np
import matplotlib.pyplot as plt


def analytical_solution(t, alpha):
    return t**alpha


def euler_method(y0, t_start, t_end, h, alpha):
    num_steps = int((t_end - t_start) / h)
    t_values = np.linspace(t_start, t_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = y0

    for i in range(num_steps):
        t_n = t_values[i]
        y_values[i + 1] = y_values[i] + h * (alpha * t_n ** (alpha - 1))
    return t_values, y_values


alphas = [2.5, 1.5, 1.1]
h_values = [0.2, 0.1, 0.05]
t_start = 0.0
t_end = 5.0

for alpha in alphas:
    plt.figure(figsize=(10, 6))

    t_analytical_plot = np.linspace(t_start, t_end, 500)
    plt.plot(
        t_analytical_plot,
        analytical_solution(t_analytical_plot, alpha),
        label="Rozwiązanie analityczne",
        color="black",
        linestyle="--",
    )

    for h in h_values:
        t_numerical, y_numerical = euler_method(
            y0=0, t_start=t_start, t_end=t_end, h=h, alpha=alpha
        )
        plt.plot(t_numerical, y_numerical, label=f"Metoda Eulera (h={h})")

    plt.title(f"Porównanie rozwiązania analitycznego i numerycznego (alpha={alpha})")
    plt.xlabel("t")
    plt.ylabel("y(t)")
    plt.legend()
    plt.grid(True)
    plt.show()
```
