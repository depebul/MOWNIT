```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt


def f_a(x):
    return x**3 - 5 * x


def df_a(x):
    return 3 * x**2 - 5


def f_b(x):
    return x**3 - 3 * x + 1


def df_b(x):
    return 3 * x**2 - 3


def f_c(x):
    return 2 - x**5


def df_c(x):
    return -5 * x**4


def f_d(x):
    return x**4 - 4.29 * x**2 - 5.29


def df_d(x):
    return 4 * x**3 - 8.58 * x


plt.figure(figsize=(16, 12))

# Plot function (a)
plt.subplot(2, 2, 1)
x_vals = np.linspace(-3, 3, 1000)
y_vals = f_a(x_vals)
plt.plot(x_vals, y_vals)
plt.axhline(y=0, color="k", linestyle="--", alpha=0.3)
plt.grid(True, alpha=0.3)
plt.title("(a) f(x) = x - 5x")
plt.xlabel("x")
plt.ylabel("f(x)")

# Plot function (b)
plt.subplot(2, 2, 2)
x_vals = np.linspace(-2, 2, 1000)
y_vals = f_b(x_vals)
plt.plot(x_vals, y_vals)
plt.axhline(y=0, color="k", linestyle="--", alpha=0.3)
plt.grid(True, alpha=0.3)
plt.title("(b) f(x) = x - 3x + 1")
plt.xlabel("x")
plt.ylabel("f(x)")

# Plot function (c)
plt.subplot(2, 2, 3)
x_vals = np.linspace(-1.5, 1.5, 1000)
y_vals = f_c(x_vals)
plt.plot(x_vals, y_vals)
plt.axhline(y=0, color="k", linestyle="--", alpha=0.3)
plt.grid(True, alpha=0.3)
plt.title("(c) f(x) = 2 - x")
plt.xlabel("x")
plt.ylabel("f(x)")

# Plot function (d)
plt.subplot(2, 2, 4)
x_vals = np.linspace(-3, 3, 1000)
y_vals = f_d(x_vals)
plt.plot(x_vals, y_vals)
plt.axhline(y=0, color="k", linestyle="--", alpha=0.3)
plt.grid(True, alpha=0.3)
plt.title("(d) f(x) = x - 4.29x - 5.29")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.tight_layout()
plt.show()


print("Prba metody Newtona i bisection jako zapasowa:")


# (a) f(x) = x^3 - 5x
print("\n(a)")
try:
    result_a = optimize.newton(f_a, 1.0, fprime=df_a)
    print(f"Newton OK: znaleziono pierwiastek x = {result_a}")
except RuntimeError as e:
    print(f"Newton FAIL: {str(e)}")
    # Try with a different starting point
    try:
        result_a = optimize.newton(f_a, 2.0, fprime=df_a)
        print(f"Newton OK z nowym punktem: znaleziono pierwiastek x = {result_a}")
    except RuntimeError as e:
        print(f"Newton FAIL z nowym punktem: {str(e)}")

# (b) f(x) = x^3 - 3x + 1
print("\n(b)")
try:
    result_b = optimize.newton(f_b, 1.0, fprime=df_b)
    print(f"Newton OK: znaleziono pierwiastek x = {result_b}")
except RuntimeError as e:
    print(f"Newton FAIL: {str(e)}")
    # Try with a different starting point
    try:
        result_b = optimize.newton(f_b, 1.3, fprime=df_b)
        print(f"Newton OK z nowym punktem: znaleziono pierwiastek x = {result_b}")
    except RuntimeError as e:
        print(f"Newton FAIL z nowym punktem: {str(e)}")

# (c) f(x) = 2 - x^5
print("\n(c)")
try:
    result_c = optimize.newton(f_c, 0.01, fprime=df_c)
    print(f"Newton OK: znaleziono pierwiastek x = {result_c}")
except RuntimeError as e:
    print(f"Newton FAIL: {str(e)}")
    # Try with a different starting point
    try:
        result_c = optimize.newton(f_c, 1, fprime=df_c)
        print(f"Newton OK z nowym punktem: znaleziono pierwiastek x = {result_c}")
    except RuntimeError as e:
        print(f"Newton FAIL z nowym punktem: {str(e)}")

# (d) f(x) = x^4 - 4.29x^2 - 5.29
print("\n(d)")
try:
    result_d = optimize.newton(f_d, 0.8, fprime=df_d)
    print(f"Newton OK: znaleziono pierwiastek x = {result_d}")
except RuntimeError as e:
    print(f"Newton FAIL: {str(e)}")
    # Try with a different starting point
    try:
        result_d = optimize.newton(f_d, 2.0, fprime=df_d)
        print(f"Newton OK z nowym punktem: znaleziono pierwiastek x = {result_d}")
    except RuntimeError as e:
        print(f"Newton FAIL z nowym punktem: {str(e)}")
```
